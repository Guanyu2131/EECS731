## Data Cleaning 1
Drop unused column from data set and make it easier to observe

```python
df4 = df1.drop(['year','latitude','longitude','region','zipCode','url'],axis='columns')
df5 = df2.drop(['year','latitude','longitude','region','zipCode','url'],axis='columns')
df6 = df3.drop(['year','latitude','longitude','region','zipCode','url'],axis='columns')
```

```python
df4.head()
```

```python
df5.head()
```

```python
df6.head()
```

## Data Cleaning 2
Remove the data with null.

```python
df4 = df4.dropna()
df5 = df5.dropna()
df6 = df6.dropna()
```

## Merging the data together
Merge one star, two stars and three stars Michelin restaurants

```python
df_append = df4.append(df5, ignore_index=True)
df_append = df_append.append(df6, ignore_index=True)
df_append.isnull().sum()
```

```python
df_append
```

## Transforming the Data
dallor sign converts to price range.The price display was broken and a lot of 'five dallors' sign only show 'one dallor', I double checked origin data. The final present is correct.  

```python
df_append.groupby('price')['price'].agg('count')
```

```python
def mkPrice(n):
    if n=='$': price = 10
    elif n=='$$': price = 50
    elif n=='$$$': price = 150
    elif n=='$$$$': price = 350
    else: price = 750
    return(price)
```

```python
df_append['price']=df_append.price.apply(mkPrice)
df_append
```

## Visualization


#### 20 Most cheapest Michelin Restaurants Worldwide

```python
df_50_low_restaurant = df_append.nsmallest(50, 'price')
df_50_low_restaurant
```

#### Plotting

```python
city = df_50_low_restaurant.groupby('city')['city'].agg('count')
plt.figure(figsize=(9,3),dpi=100)
city.plot()
plt.title("50 Cheapest Michelin Restaurants Location")
plt.xlabel('city')
plt.ylabel('numbers')
```

```python
cuisine = df_50_low_restaurant.groupby('cuisine')['cuisine'].agg('count')
plt.figure(figsize=(9,3),dpi=100)
cuisine.plot()
plt.title("20 Cheapest Michelin Restaurants Species")
plt.xlabel('cuisines')
plt.ylabel('numbers')
```
